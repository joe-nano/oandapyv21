## 円建てmargin計算

```python
# 基底通貨が円以外のペア
not_yen_pair = inst_df[inst_df['instrument'].str.split('_', expand=True)[1]!='JPY'].drop_duplicates(subset='instrument')['instrument'].values
```

```python
# 各ペアの円建てレートを取得
yen_rate = {}
for p in not_yen_pair:
    try:
        inst = p.split('_')[1] + '_JPY'
        price = mean(get_price(inst).values())
        yen_rate[inst] = price
    except V20Error:
        inst = 'JPY_' + p.split('_')[1]
        price = mean(get_price(inst).values())
        yen_rate[inst] = price
        
print('取得時間: {}'.format(datetime.datetime.now()))
```

```python
# 円建てmargin計算
inst_df2 = pd.DataFrame()
for y in yen_rate.items():
    if y[0].split('_')[1] == 'JPY':
        df = inst_df[inst_df['instrument'].str.split('_', expand=True)[1]==y[0].split('_')[0]].copy()
        df['margin(円)'] = df['profit'] * y[1]
    else:
        df = inst_df[inst_df['instrument'].str.split('_', expand=True)[1]==y[0].split('_')[1]].copy()
        df['margin(円)'] = df['profit'] / y[1]
    inst_df2 = pd.concat([inst_df2, df], axis=0)
df = inst_df[inst_df['instrument'].str.split('_', expand=True)[1]=='JPY'].copy()
df['margin(円)'] = df['profit']
inst_df2 = pd.concat([inst_df2, df], axis=0)
inst_df2.sort_index(inplace=True)
```
